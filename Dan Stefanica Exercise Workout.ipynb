Use the Black-Scholes formula to price a six months European call option with strike 40, on an underlying asset with spot price 42 and volatility 30%, which pays dividends continuously, with dividend rate 3%. Assume that interest rates are constant at 5%.

```python
import math
from scipy.stats import norm
K=40
S=42
V=0.3
q=0.03
r=0.05
t=0.5
#C=Sexp(-qt)N(d1)-Kexp(-rt)N(d2)
#d1=log(S/K)+(r-q+V^2/2)t / V sqrt(t)
srt=V*t**0.5
d1=(math.log(S/K)+(r-q+V**2/2)*t)/srt
d2=d1-srt
C=S*math.exp(-q*t)*norm.cdf(d1)-K*math.exp(-r*t)*norm.cdf(d2)
print("Call Option price is {}".format(C))
```

Compute the expected value and the variance of the stock price at time T = 3T, i.e., compute E[ST] and var(ST).

```python
from math import factorial as fact
import numpy as np
C=lambda n,r:fact(n)/fact(r)/fact(n-r)
s=40
u=1.05
d=0.96
pu=0.6
pd=0.4
n=3
S=np.array([s*u**(n-i)*d**i for i in range(n+1)])
P=np.array([C(n,i)*pu**(n-i)*pd**i for i in range(n+1)])
ES=sum(S*P)
print("Expected Stock Price at 3T is {}".format(ES))
ES2=sum(S*S*P)
VS=ES2-ES*ES
print("Variance of Stock Price at 3T is {}".format(VS))
```

**Newton's Method**  
$f(\sigma)=Se^{-qT}N(d_1)-Ke^{-rT}N(d_2)-C$  
$f'(\sigma)=Se^{-qT}N'(d_1)\left(\frac{dd_1}{d\sigma}-\frac{dd_2}{d\sigma}\right)$  
$\frac{dd_1}{d\sigma}-\frac{dd_2}{d\sigma}=\sqrt{T}$  
$f'(\sigma)=Se^{-qT}\sqrt{T}N'(d_1)$  
$f'(\sigma)=Se^{-qT}\sqrt{\frac{T}{2\pi}}e^{-\frac{d_1^2}{2}}$
