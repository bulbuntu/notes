$dS= rSdt+\sigma SdW$  
$\qquad\qquad$Where $dW=\sqrt t Z dt$ is a Weiner process.  
The solution for this SDE is  

<div style="border:dotted;text-align:center;word-wrap:break-word">$log(\frac{S_T}{S_0})=(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt T Z$

Now for Call Option Value $C_T$ at Maturity $T$ is defined as: 

$C_T=$  
$\qquad\qquad=S_T-K$  if $S_T\geq K$  
$\qquad\qquad=0$  otherwise  

**For $S_T\geq K$**  
$S_T \geq K \implies S_0e^{(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt TZ} \geq K$  
$\quad\implies(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt TZ \geq log(\frac{K}{S_0})$  
$\quad\implies\sigma\sqrt TZ \geq log(\frac{K}{S_0})-(r-q-\frac{\sigma^2}{2})T$  
$\quad\implies Z \geq\frac{log(\frac{K}{S_0})-(r-q-\frac{\sigma^2}{2})T}{\sigma\sqrt T}$  
$\quad\implies Z \geq-\frac{log(\frac{S_0}{K})+(r-q-\frac{\sigma^2}{2})T}{\sigma\sqrt T}$  
$\quad\implies Z \geq-d_2$  

Now Rewriting Call Option Value  
$C_T=$  
$\qquad\qquad=S_T-K$  if $Z\geq -d_2$  
$\qquad\qquad=0$  otherwise  

Discount for Call Option price at t=0  
$C_0=$  
$\qquad\qquad=e^{-rT}(S_T-K)$  if $Z\geq -d_2$  
$\qquad\qquad=0$  otherwise

$\qquad\qquad=e^{-rT}(S_0e^{(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt T Z}-K)$  if $Z\geq -d_2$

9) Consider a six months plain vanilla European call option with strike 18 on a non-dividend-paying underlying asset with spot price 20. Assume that the asset has lognormal distribution with volatility 20% and that interest rates are constant at 5%.  
(i) Compute the Greeks of the call option, i.e, $\Delta, \, \Gamma,\, \rho,\, vega\,\,\, and\,\,\, \Theta$.  
Use finite differences to find approximate values for the Greeks.  
Denote by C(S, K, T, $\sigma$, r) the value of the call option obtained from the Black-Scholes formula.

```python
from scipy.stats import norm
import math
K=18
S=20
v=0.2
r=0.05
T=0.5
d1=lambda S,K,r,v,T:(math.log(S/K)+(r+v**2/2)*0.5)/v/T**0.5
d2=lambda S,K,r,v,T:(math.log(S/K)+(r-v**2/2)*0.5)/v/T**0.5
C=lambda S,K,r,v,T:S*norm.cdf(d1(S,K,r,v,T))-K*math.exp(-r*T)*norm.cdf(d2(S,K,r,v,T))
CV=C(S,K,r,v,T)
delta=(C(S+0.01*S,K,r,v,T)-CV)/0.01/S
gamma=(C(S+0.01*S,K,r,v,T)-2*CV+C(S-0.01*S,K,r,v,T))/0.01**2/S**2
rho=(C(S,K,r+0.01*r,v,T)-CV)/0.01/r
vega=(C(S,K,r,v+0.001,T)-CV)/0.001
theta=(CV-C(S,K,r,v,T-0.01*T))/0.01/T
print("Delta={:.4f}".format(delta))
print("Gamma={:.4f}".format(gamma))
print("Rho={:.4f}".format(rho))
print("Vega={:.4f}".format(vega))
print("Theta={:.4f}".format(theta))

#BSDelta
bsdelta=norm.cdf(d1(S,K,r,v,T))
print("BS Delta={:.4f}".format(bsdelta))
#Gamma
#1/v/sqrt(T)/S
bsgamma=math.exp(-d1(S,K,r,v,T)**2/2)/v/S/(2*math.pi*T)**0.5
print("BS Gamma={:.4f}".format(bsgamma))
bsrho=K*T*math.exp(-r*T)*norm.cdf(d2(S,K,r,v,T))
print("BS Rho={:.4f}".format(bsrho))
bsvega=S*T**0.5/(2*math.pi)**0.5*math.exp(-d1(S,K,r,v,T)**2/2)
print("BS Vega={:.4f}".format(bsvega))
bstheta=S*v/2/(2*math.pi*T)**0.5*math.exp(-d1(S,K,r,v,T)**2/2.) + r*K*math.exp(-r*T)*norm.cdf(d2(S,K,r,v,T))
print("BS Theta={:.4f}".format(bstheta))
```
