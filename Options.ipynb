### Deriving Black Scholes Formula

$\frac{S(t+\delta t)-S(t)}{S(t)\delta t}=r+\frac{\sigma}{\sqrt{\delta t}}Z$  
$dS= rSdt+\sigma SdW$  
$\qquad\qquad$Where $dW=\sqrt t Z dt$ is a Weiner process.  
The solution for this SDE is  

<div style="border:dotted;text-align:center;word-wrap:break-word">$log(\frac{S_T}{S_0})=(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt T Z$  
$S_T=S_0e^{(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt T Z}$  
</div>
Now for Call Option Value $C_T$ at Maturity $T$ is defined as: 

$C_T=$  
$\qquad\qquad=S_T-K$  if $S_T\geq K$  
$\qquad\qquad=0$  otherwise  

**For $S_T\geq K$**  
$S_T \geq K \implies S_0e^{(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt TZ} \geq K$  
$\quad\implies(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt TZ \geq log(\frac{K}{S_0})$  
$\quad\implies\sigma\sqrt TZ \geq log(\frac{K}{S_0})-(r-q-\frac{\sigma^2}{2})T$  
$\quad\implies Z \geq\frac{log(\frac{K}{S_0})-(r-q-\frac{\sigma^2}{2})T}{\sigma\sqrt T}$  
$\quad\implies Z \geq-\frac{log(\frac{S_0}{K})+(r-q-\frac{\sigma^2}{2})T}{\sigma\sqrt T}$  
$\quad\implies Z \geq-d_2$  

Now Rewriting Call Option Value  
$C_T=$  
$\qquad\qquad=S_T-K$  if $Z\geq -d_2$  
$\qquad\qquad=0$  otherwise  

Discount for Call Option price at t=0  
$C_0=$  
$\qquad\qquad=e^{-rT}(S_T-K)$  if $Z\geq -d_2$  
$\qquad\qquad=0$  otherwise  

$C_0=$  
$\qquad\qquad=e^{-rT}(S_0e^{(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt T Z}-K)$  if $Z\geq -d_2$  
$\qquad\qquad=0$  otherwise  

Now evaluating only the non-zero part of price of Call Option  
$C_0=\frac{e^{-rT}}{\sqrt{2\pi}}\int_{-d_2}^{\infty}{(S_0e^{(r-q-\frac{\sigma^2}{2})T+\sigma\sqrt T x}-K) e^{-\frac{x^2}{2}}\,dx}$  
$\qquad=\frac{S_0e^{-qT}}{\sqrt{2\pi}}\int_{-d_2}^{\infty}e^{-\frac{\sigma^2}{2}T+\sigma\sqrt Tx-\frac{x^2}{2}}dx-\frac{Ke^{-rT}}{\sqrt{2\pi}}\int_{-d_2}^{\infty}e^{-\frac{x^2}{2}}dx$  
$\qquad=\frac{S_0e^{-qT}}{\sqrt{2\pi}}\int_{-d_2}^{\infty}e^{-\frac{(x-\sigma\sqrt T)^2}{2}}dx-Ke^{-rT}N(d_2)\qquad$Now substitute $t=x-\sigma\sqrt T$    
$\qquad=\frac{S_0e^{-qT}}{\sqrt{2\pi}}\int_{-d_2-\sigma\sqrt T}^{\infty}{e^{-\frac{t^2}{2}}\,dt}-Ke^{-rT}N(d_2)\qquad$ As $d_1=d_2+\sigma\sqrt T$    
$\qquad=\frac{S_0e^{-qT}}{\sqrt{2\pi}}\int_{-d_1}^{\infty}e^{-\frac{t^2}{2}}dt-Ke^{-rT}N(d_2)\qquad$  
$\qquad=S_0e^{-qT}N(d_1)-Ke^{-rT}N(d_2)\qquad$  



**Relationship of a delta neutral portfolio to $\Theta$, $\Gamma$ and $\sigma$**  

Taylor Expansion of Option Value V(S,T)  

$V(S+dS,T+dt) = V(S,T) +dS\frac{\delta V(S,T)}{\delta S} + dt\frac{\delta V(S,T)}{\delta t} + \frac{(dS)^2}{2}\frac{\delta^2 V(S,T)}{\delta S^2} + \frac{(dt)^2}{2}\frac{\delta^2 V(S,T)}{\delta t^2} +dS\,dt\frac{\delta^2V(S,T)}{\delta S\delta t} + O((dS)^3) + O((dt)^3)$  
$V(S+dS,T+dt) \approx V(S,T) +dS\frac{\delta V(S,T)}{\delta S} + dt\frac{\delta V(S,T)}{\delta t} + \frac{(dS)^2}{2}\frac{\delta^2 V(S,T)}{\delta S^2} $Ignoring the higher order dt  
$V(S+dS,T+dt) \approx V(S,T) +dS\Delta + \Theta dt + \frac{(dS)^2}{2}\Gamma $  
$dV-\Delta dS\approx \Theta dt + \frac{(dS)^2}{2}\Gamma $  

Remember the Log Normality assumption for Stock Price  
$dS=rSdt+\sigma S Z \sqrt dt$  
Thus  
$(dS)^2=r^2S^2(dt)^2 + 2rS^2Zdt\sqrt dt + \sigma^2S^2Z^2dt$If we ignore dt of the order>1  
$(dS)^2\approx \sigma^2S^2Z^2dt$  
$E[Z^2]=1$ as Z is a SNV  
$(dS)^2\approx \sigma^2S^2dt$  
Thus  
$dV-\Delta dS\approx \Theta dt + \frac{\sigma^2S^2}{2}\Gamma\,dt $  
$d\Pi\approx \Theta dt + \frac{\sigma^2S^2}{2}\Gamma\,dt\qquad \Pi$ being the Delta Neutral portfolio  
If we assume the $\Pi$ grows at the rate of risk free rate $r$  

$d\Pi=r\Pi dt\approx\Theta dt + \frac{\sigma^2S^2}{2}\Gamma\,dt$  

Finally  
<div style="border:dotted;text-align:center">$r\Pi\approx\Theta  + \frac{\sigma^2S^2}{2}\Gamma$</div>
