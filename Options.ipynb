### Black Scholes

* Put Call Parity  
$P+Se^{-q(T-t)}-C=Ke^{-r(T-t)}$  
* Call Option Price(at time $t$ with expiry=$T$  
$C=Se^{-q(T-t)}N(d1)-Ke^{-r(T-t)}N(d2)$  
$d1=\frac{ln(\frac{S}{K})+(r-q+\frac{\sigma^2}{2})(T-t)}{\sigma\sqrt{T-t}}$  
$d2=d1-\sigma\sqrt {T-t}$  
$\quad=\frac{ln(\frac{S}{K})+(r-q-\frac{\sigma^2}{2})(T-t)}{\sigma\sqrt{T-t}}$  
* Put Option Price(at current time 0 and expiry=$t$)  
$P=Ke^{-rt}-Se^{-qt}+C$  
$\quad=Ke^{-rt}-Se^{-qt}+Se^{-qt}N(d1)-Ke^{-rt}N(d2)$  
$\quad=Ke^{-rt}(1-N(d2))-Se^{-qt}(1-N(d1))$  
$\quad=Ke^{-rt}N(-d2)-Se^{-qt}N(-d1)$  

* Option Delta  
$\Delta_C=e^{-q(T-t)}N(d_1)$  
$\Delta_P=-e^{-q(T-t)}N(-d_1)$  
* Option Gamma  
$\Gamma_C=\Gamma_P=\frac{e^{-q(T-t)}e^{-\frac{d_1^2}{2}}}{S\sigma\sqrt{T-t}}$  
* Option Vega  
$\kappa_C=\kappa_P=\frac{Se^{-qt}e^{-\frac{d_1^2}{2}}\sqrt t}{\sqrt{2\pi}}$   
* Option Rho  
$\rho_C=(T-t)Ke^{-r(T-t)}N(d_2)$  
$\rho_P=-(T-t)Ke^{-r(T-t)}N(-d_2)$  
* Option Theta  
$\Theta_C=qSe^{-q(T-t)}N(d_1)-rKe^{-r(T-t)}N(d_2)-\frac{S\sigma e^{-q(T-t)} e^{-\frac{d_1^2}{2}}}{2\sqrt{2\pi(T-t)}}$  
$\Theta_P=-qSe^{-q(T-t)}N(-d_1)+rKe^{-r(T-t)}N(-d_2)-\frac{S\sigma e^{-q(T-t)} e^{-\frac{d_1^2}{2}}}{2\sqrt{2\pi(T-t)}}$  


**Note for future derivations**  
$Se^{-q(T-t)}=Ke^{-r(T-t)}\frac{S}{K}e^{(r-q)(T-t)}$  
$\quad=Ke^{-r(T-t)}\operatorname{exp}( log(\frac{S}{K}) +(r-q)(T-t))$  
$\qquad\qquad log(\frac{S}{K}) +(r-q)(T-t)=log(\frac{S}{K}) +(r-q+\frac{\sigma^2}{2})(T-t) -\frac{\sigma^2}{2}(T-t)$  
$\qquad\qquad log(\frac{S}{K}) +(r-q)(T-t)=d_1\sigma\sqrt{T-t} -\frac{\sigma^2}{2}(T-t)$  
$Se^{-q(T-t)}=Ke^{-r(T-t)}\operatorname{exp}(d_1\sigma\sqrt{T-t} -\frac{\sigma^2}{2}(T-t))$  
Multiply both sides by $e^{-\frac{d_1^2}{2}}$  
$Se^{-q(T-t)}e^{-\frac{d_1^2}{2}}=Ke^{-r(T-t)}\operatorname{exp}(d_1\sigma\sqrt{T-t} -\frac{\sigma^2}{2}(T-t)-\frac{d_1^2}{2})$  
$\quad=Ke^{-r(T-t)}\operatorname{exp}(-\frac{(d_1-\sigma\sqrt{T-t})^2}{2})$  
$\quad=Ke^{-r(T-t)}\operatorname{exp}(-\frac{d_2^2}{2})$  
$Se^{-q(T-t)}e^{-\frac{d_1^2}{2}}=Ke^{-r(T-t)}e^{-\frac{d_2^2}{2}}$   
$Se^{-q(T-t)}N'(d_1)=Ke^{-r(T-t)}N'(d_2)$  

**Option Delta Derivation**  
$C=Se^{-qt}N(d_1)-Ke^{-rt}N(d_2)$  
$\frac{dC}{dS}=e^{-qt}N(d1) + Se^{-qt}N'(d_1)\left(\frac{dd_1}{dS}-\frac{dd_2}{dS}\right)$  
$\qquad\qquad$ From earlier $Se^{-qt}N'(d_1)=Ke^{-rt}N'(d_2)$   
$\frac{dd_1}{dS}=\frac{1}{\sigma\sqrt{T-t}}\frac{K}{S}\frac{1}{K}=\frac{1}{S\sigma\sqrt{T-t}}=\frac{dd_2}{dS}$  
Thus  
$\Delta_C=e^{-q(T-t)}N(d_1)$  
$\Delta_P=-e^{-q(T-t)}N(-d_1)\qquad$Could be derived from put call parity equation  
**Option Gamma Derivation**  
$\frac{d\Delta_C}{dS}=e^{-q(T-t)}N'(d_1)\frac{1}{S\sigma\sqrt{T-t}}$  
$\frac{d\Delta_C}{dS}=\frac{e^{-q(T-t)}e^{-\frac{d_1^2}{2}}}{S\sigma\sqrt{T-t}}$  
$\Gamma_C=\Gamma_P=\frac{e^{-q(T-t)}e^{-\frac{d_1^2}{2}}}{S\sigma\sqrt{T-t}}$  
**Option Vega Derivation**  
$\frac{dC}{d\sigma}=Se^{-qt}N'(d_1)\left(\frac{dd_1}{d\sigma}-\frac{dd_2}{d\sigma}\right)$  
$\frac{dd_1}{d\sigma}=-\frac{log\frac{S}{K}+(r-q+\sigma^2/2)t}{\sigma^2\sqrt{t}} +\frac{\sigma t}{\sigma\sqrt t}$
$\qquad=\frac{A}{\sigma^2\sqrt t} - \frac{t\sigma^2/2}{\sigma^2\sqrt{t}}+\frac{\sigma t}{\sigma\sqrt t}$  
$\qquad=\frac{A}{\sigma^2\sqrt t} +\frac{\sqrt t}{2}$  
$\frac{dd_1}{d\sigma}-\frac{dd_2}{d\sigma}=\frac{A}{\sigma^2\sqrt t} +\frac{\sqrt t}{2} -(\frac{A}{\sigma^2\sqrt t} -\frac{\sqrt t}{2})$  
$\frac{dd_1}{d\sigma}-\frac{dd_2}{d\sigma}=\sqrt t$  
$\frac{dC}{d\sigma}=Se^{-qt}N'(d_1)\sqrt t$  
$\frac{dC}{d\sigma}=\frac{Se^{-qt}e^{-\frac{d_1^2}{2}}\sqrt t}{\sqrt{2\pi}}$   
$\kappa_C=\frac{Se^{-qt}e^{-\frac{d_1^2}{2}}\sqrt t}{\sqrt{2\pi}}$   
**Option Rho Derivation**  
$\frac{dC}{dr}=(T-t)Ke^{-r(T-t)}N(d_2) +Se^{-q(T-t)}N'(d_1)\left(\frac{dd_1}{dr}-\frac{dd_2}{dr}\right)$  
$\frac{dd_1}{dr}=\frac{T-t}{\sigma\sqrt{T-t}}$  
$\frac{dd_1}{dr}-\frac{dd_2}{dr}=0$  
$\frac{dC}{dr}=-(T-t)Ke^{-r(T-t)}N(d_2)$  
$\rho_C=-(T-t)Ke^{-r(T-t)}N(d_2)$  
**Option Theta Derivation**  
$\frac{dC}{dt}=qSe^{-q(T-t)}N(d_1)-rKe^{-r(T-t)}N(d_2)+Se^{-q(T-t)}N'(d_1)\left(\frac{dd_1}{dt}-\frac{dd_2}{dt}\right)$  
$\frac{dd_1}{dt}=\frac{1}{2}\frac{log\frac{S}{K}+(r-q+\sigma^2/2)(T-t)}{\sigma(T-t)\sqrt{T-t}}-\frac{r-q+\sigma^2/2}{\sigma\sqrt{T-t}}$  
$\qquad=A+\frac{1}{2}\frac{\sigma^2/2}{\sigma\sqrt{T-t}}-\frac{\sigma^2/2}{\sigma\sqrt{T-t}}$  
$\qquad=A-\frac{\sigma}{4\sqrt{T-t}}$  
$\frac{dd_1}{dt}-\frac{dd_2}{dt}=A-\frac{\sigma}{4\sqrt{T-t}}-(A+\frac{\sigma}{4\sqrt{T-t}})$  
$\frac{dd_1}{dt}-\frac{dd_2}{dt}=-\frac{\sigma}{2\sqrt{T-t}}$  
$\frac{dC}{dt}=qSe^{-q(T-t)}N(d_1)-rKe^{-r(T-t)}N(d_2)-Se^{-q(T-t)}N'(d_1)\frac{\sigma}{2\sqrt{T-t}}$  
$\frac{dC}{dt}=qSe^{-q(T-t)}N(d_1)-rKe^{-r(T-t)}N(d_2)-\frac{S\sigma e^{-q(T-t)} e^{-\frac{d_1^2}{2}}}{2\sqrt{2\pi(T-t)}}$  
$\Theta_C=qSe^{-q(T-t)}N(d_1)-rKe^{-r(T-t)}N(d_2)-\frac{S\sigma e^{-q(T-t)} e^{-\frac{d_1^2}{2}}}{2\sqrt{2\pi(T-t)}}$
